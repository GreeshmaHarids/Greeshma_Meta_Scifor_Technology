## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/UniversalBank.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data.drop(["ID", "ZIP Code", "Family","CCAvg","Education","Mortgage","Securities Account","CD Account"],axis=1,inplace=True)
```

```python
X=data.drop('Personal Loan',axis=1)
y=data['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## Prediction
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['linear']
              }
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
print(accuracy_score(y_test,y_pred))
```
